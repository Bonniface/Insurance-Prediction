```python
# Check if there any missing values in train set
ax = Train.isna().sum().sort_values().plot(kind = 'barh', figsize = (10, 7))
plt.title('Percentage of Missing Values Per Column in Train Set', fontdict={'size':15})
for p in ax.patches:
    percentage ='{:,.0f}%'.format((p.get_width()/Train.shape[0])*100)
    width, height =p.get_width(),p.get_height()
    x=p.get_x()+width+0.02
    y=p.get_y()+height/2
    ax.annotate(percentage,(x,y))
```

```python
# Removes Data Duplicates while Retaining the First one
def remove_duplicate(data):
    data.drop_duplicates(keep="first", inplace=True)
    return "Checked Duplicates"

# Removes Duplicates from train data if any
remove_duplicate(Train)
```

There are numerous issues with the data that have to be dealt with prior  model building. The first thing we will be taking care of is converting categorical columns into dummy variables , filling the missing values.

```python
CAT = ['Building_Painted', 'Building_Fenced', 'Garden',
        'NumberOfWindows', 'Geo_Code']
```

```python
from sklearn.impute import SimpleImputer

for col in CAT:
    cat_imputer = SimpleImputer(strategy='most_frequent')
    cat_imputer.fit(Train[[col]])
    Train.loc[:, col] = cat_imputer.transform(Train[[col]])
    Test.loc[:, col] = cat_imputer.transform(Test[[col]])
```

```python
Train['Building_Fenced'] = Train['Building_Fenced'].map({'V':0,'N':1})
Train['Building_Painted'] = Train['Building_Painted'].map({'V':0,'N':1})
Train['Garden'] = Train['Garden'].map({'O':0,'V':1})
Train['Settlement'] = Train['Settlement'].map({'U':0,'R':1})

```

```python
Trend = Train[['YearOfObservation','Residential','Building_Painted', 'Building_Fenced']].groupby('YearOfObservation').agg([np.sum])

Trend
```

We can see a downward trend in Residential buildings being insured as well as painted and fenced buildings being insured. This is likely due to the former drop in residential buildings being insured or the Insurance firm taking on more risk.
